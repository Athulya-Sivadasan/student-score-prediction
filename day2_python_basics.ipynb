```python
# Variables and Data Types
name = "Athulya"          # text (string)
age = 21                  # whole number (integer)
height = 5.4               # decimal number (float)
is_student = True          # True/False (boolean)

print(name, age, height, is_student)
print(type(name), type(age), type(height), type(is_student))
```

```python
# Operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Is a greater than b?", a > b)
```

```python
# Loops
for i in range(1, 6):
    print("Study hour:", i)
```

```python
# Functions
def greet_student(student_name):
    return f"Hello, {student_name}! Ready to study?"

message = greet_student("Athulya")
print(message)
```

```python
# Simple program: study hours tracker
def predict_effort(hours):
    if hours < 2:
        return "Low effort"
    elif hours < 5:
        return "Moderate effort"
    else:
        return "High effort"

study_days = [1, 3, 6, 2, 8]

for hours in study_days:
    print(f"{hours} hours -> {predict_effort(hours)}")
```
