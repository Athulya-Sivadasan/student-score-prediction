Today I learned NumPy arrays, indexing, slicing, and basic math operations.

## Step 1: Import NumPy

```python
import numpy as np
```

## Step 2: Create an Array
An array is like a list of numbers. Here we store study hours for each day of the week.

```python
study_hours = np.array([2, 3, 4, 1, 5, 6, 2])
print(study_hours)
print(type(study_hours))
```

## Step 3: Indexing
Indexing means picking one value out of the array. Counting starts at 0.

```python
print(study_hours[0])   # first value
print(study_hours[3])   # fourth value
print(study_hours[-1])  # last value
```

## Step 4: Slicing
Slicing means picking a range of values from the array.

```python
print(study_hours[0:3])   # first 3 values
print(study_hours[3:])    # from index 3 to the end
```

## Step 5: Basic Math Operations
NumPy lets us do math on the whole array at once, without loops.

```python
print(study_hours + 1)      # adds 1 to every value
print(study_hours * 2)      # doubles every value
print(study_hours.sum())    # total hours
print(study_hours.mean())   # average hours per day
print(study_hours.max())    # highest value
print(study_hours.min())    # lowest value
```

## Step 6: 2D Arrays
Real datasets often have rows and columns, like a spreadsheet. Each row here is [study_hours, score].

```python
data = np.array([[2, 65], [3, 70], [4, 75], [5, 85]])
print(data)
print(data.shape)   # (rows, columns)
```
