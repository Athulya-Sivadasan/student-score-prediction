```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_scores_clean.csv")
df.head()
```

```python
plt.scatter(df["Hours"], df["Scores"], color="blue")
plt.title("Study Hours vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
plt.bar(range(1, 11), df["Scores"][:10], color="green")
plt.title("Scores of First 10 Students")
plt.xlabel("Student Number")
plt.ylabel("Score")
plt.show()
```

```python
plt.plot(df["Scores"][:15], marker="o", color="orange")
plt.title("Score Trend (First 15 Students)")
plt.xlabel("Student Index")
plt.ylabel("Score")
plt.show()
```
