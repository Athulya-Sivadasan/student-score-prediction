```python
study_hours = np.array([2, 3, 4, 1, 5, 6, 2])
print(study_hours)
print(type(study_hours))
```

```python
print(study_hours[0])   # first value -> 2
print(study_hours[3])   # fourth value -> 1
print(study_hours[-1])  # last value -> 2
```

```python
print(study_hours[0:3])   # first 3 values -> [2 3 4]
print(study_hours[3:])    # from index 3 to the end -> [1 5 6 2]print(study_hours + 1)      # adds 1 to every value
print(study_hours * 2)      # doubles every value
print(study_hours.sum())    # adds up all values -> total hours
print(study_hours.mean())   # average hours per day
print(study_hours.max())    # highest value
print(study_hours.min())    # lowest value
```
